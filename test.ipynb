```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://finance.yahoo.com/markets/world-indices/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
page = requests.get(URL, headers=headers)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find(id="svelte")
```

```python
symbol = results.find_all("span", class_="symbol yf-90gdtp")
```

```python
name = results.find_all("div", class_="leftAlignHeader companyName yf-362rys enableMaxWidth")
```

```python
price = results.find_all("fin-streamer")
```

```python
for i in range(len(symbol)):
    print(name[i].text.strip() + "(" + symbol[i].text.strip() + "): " +  price[i*6].text.strip())
```
